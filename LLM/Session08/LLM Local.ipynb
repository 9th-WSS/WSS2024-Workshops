# Imports

````python
from pynight.common_openai import (
    print_chat_streaming,
)
````

````python
from pynight.common_icecream import ic

from llama_cpp import Llama
````

# Download the model

````python
model_name_or_path = "TheBloke/Starling-LM-7B-alpha-GGUF"
model_basename = "starling-lm-7b-alpha.Q4_K_M.gguf"
````

````python
from huggingface_hub import hf_hub_download

model_path = hf_hub_download(repo_id=model_name_or_path, filename=model_basename)
````

````python
model_path
````

# Load the model

````python
if cuda_p:
  n_gpu_layers = -1 #: GPU
else:
  n_gpu_layers = 0 #: CPU

llm = Llama(
    model_path=model_path,
    n_ctx=1024,
    n_gpu_layers=n_gpu_layers,
    logits_all=True,
)
````

# Using the Model

````python
output = llm(
    "Q: Name the planets in the solar system?\nA: ",
    max_tokens=512,
    stop=["\n"],
    echo=True,
    temperature=0,
    stream=True,
)

print_chat_streaming(output)
````

````python
output = llm.create_chat_completion(
    messages=[
        {"role": "system", "content": """You are a senior assistant. You answer concisely."""},
        {"role": "user", "content": """How many planets are there in the solar system?"""},
        {"role": "assistant", "content": """There are eight planets in the solar system."""},
        {"role": "user", "content": """What are the names of the planets in the solar system?"""},
    ],
    max_tokens=2048,
    stop=[],
    temperature=0,
    stream=True,
)

print_chat_streaming(output)
````

````python
output = llm(
    "Q: Name the planets in the solar system?\nA: ",
    max_tokens=512,
    stop=["\n"],
    echo=True,
    logprobs=True,
    temperature=0,
    stream=False,
)

ic(output)
None
````

## Usage

````python
! llm -m starling-7b 'Who is the Queen of England?'
````

The following options are available:

-   `-o verbose 1` - output more verbose logging
-   `-o max_tokens 100` - max tokens to return. Defaults to 4000.
-   `-o no_gpu 1` - remove the default \`n~gpulayers~=1\`\` argument, which should disable GPU usage
-   `-o n_gpu_layers 10` - increase the `n_gpu_layers` argument to a higher value (the default is `1`)
-   `-o n_ctx 1024` - set the `n_ctx` argument to `1024` (the default is `4000`)

````python
! llm -m starling-7b 'Who is the Queen of England?' \
  -o n_gpu_layers -1 \
  -o max_tokens 25
````

````python
! echo "My name is Azad." | \
  llm -m starling-7b "What is my name?" \
  -o n_gpu_layers -1 \
  -o max_tokens 25
````

We can also continue a conversation:

````python
! llm -m starling-7b "How do I write red in CSS?" \
  -o n_gpu_layers -1 \
  -o max_tokens 100
````

````python
! llm --continue "How do I write it in RGB?" \
  -o n_gpu_layers -1 \
  -o max_tokens 100
````
